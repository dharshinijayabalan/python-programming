```python
tasks = []

def show_task():
    if not tasks:
        print("No tasks in the list.")
    else:
        for i, task in enumerate(tasks, 1):
            print(f"{i}. {task}")

def add_new_task():
    task = input("Enter a task: ")
    tasks.append(task)
    print("Task added.")

def remove_task():
    show_task()
    task_no = int(input("Enter task number to remove: "))
    if 0 < task_no <= len(tasks):
        tasks.pop(task_no - 1)
        print("Task removed.")
    else:
        print("Invalid task number.")

while True:
    print("\n--- To-Do List Menu ---")
    print("1. Add New Task")
    print("2. Show Task")
    print("3. Remove Task")
    print("4. Exit")
    choice = input("Choose an option: ")

    if choice == '1':
        add_new_task()
    elif choice == '2':
        show_task()
    elif choice == '3':
        remove_task()
    elif choice == '4':
        print("Have a nice day!")
        break
    else:
        print("Invalid choice.")
```
