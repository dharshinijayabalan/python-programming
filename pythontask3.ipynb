```python
print("Welcome to Rock-Paper-Scissors by Dharshini!")
import random

def player_choice():
    print("\nChoose any one: rock, paper, or scissors")
    choice = input("Enter your choice: ")
    while choice not in ['rock', 'paper', 'scissors']:
        print("Invalid choice. Please choose rock, paper, or scissors.")
        choice = input("Enter your choice: ")
    return choice

def computer_choice():
    return random.choice(['rock', 'paper', 'scissors'])

def determine_winner(user, computer):
    if user == computer:
        return "tie"
    elif (user == 'rock' and computer == 'scissors') or \
         (user == 'scissors' and computer == 'paper') or \
         (user == 'paper' and computer == 'rock'):
        return "player"
    else:
        return "computer"

def display_result(player, computer, winner):
    print(f"\n You choose: {player}")
    print(f"Computer choose: {computer}")
    if winner == "tie":
        print("It's a tie! Please try again")
    elif winner == "player":
        print("You win! ")
    else:
        print("Computer wins. ")

def play_game():
    player_score = 0
    computer_score = 0

    while True:
        player = player_choice()
        computer = computer_choice()
        winner = determine_winner(player, computer)

        display_result(player, computer, winner)

        if winner == "player":
            player_score += 1
        elif winner == "computer":
            computer_score += 1

        print(f"\n Score => You: {player_score} | Computer: {computer_score}")

        play_again = input("\n Do you want to play again? (yes/no): ")
        if play_again != "yes":
            print("\n Thanks for playing! Final Score:")
            print(f"You: {player_score} | Computer: {computer_score}")
            break

play_game()
```
